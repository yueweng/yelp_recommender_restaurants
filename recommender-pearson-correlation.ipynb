### Get Data

```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.decomposition import TruncatedSVD

restaurants_df = pd.read_csv('data/restaurants.csv')
restaurants_df.drop(columns=['Unnamed: 0'], inplace=True)
restaurants_df.head()
```

```python
restaurants_df.info()
```

```python
user_reviews_df = pd.read_csv('data/user_reviews.csv')
user_reviews_df.drop(columns=['Unnamed: 0'], inplace=True)
user_reviews_df
```

```python
import ast
user_reviews_df['user_id'] = user_reviews_df['user'].apply(lambda x: ast.literal_eval(x)['id'])
user_reviews_df['slug'] = user_reviews_df['url'].apply(lambda x: x.split("/biz/")[1].split("?")[0])
user_reviews_df
```

```python
restaurants_name_df = restaurants_df[['alias', 'name']]
restaurants_name_df
```

```python
reviews_df = user_reviews_df[['user_id', 'slug', 'rating']]
reviews_df
```

```python
ratings_df = reviews_df.merge(restaurants_name_df, left_on='slug', right_on='alias')
ratings_df.drop(columns=['slug', 'alias'], inplace=True)
ratings_df
```

### Building a Utility Matrix

```python
def utility_matrix():
    ratings_crosstab = ratings_df.pivot_table(values='rating', index='user_id', columns='name', fill_value=0)
    return ratings_crosstab
```

### Transposing the Matrix

```python
ratings_crosstab.shape
```

```python
def transpose_matrix(ratings_crosstab):
    return ratings_crosstab.values.T
```

### Decomposing the Matrix

```python
def svd_fit_transform(X):
    SVD = TruncatedSVD(n_components=12, random_state=17)
    resultant_matrix = SVD.fit_transform(X)
    return resultant_matrix
```

### Generating a Correlation Matrix

```python
def corr_matrix(resultant_matrix):
    corr_mat = np.corrcoef(resultant_matrix)
    return corr_mat
```

### Isolate Benu from Correlation Matrix

```python
restaurant_names = ratings_crosstab.columns
restaurant_list = list(restaurant_names)

benu = restaurant_list.index('Benu')
print(restaurant_names)
```

```python
def get_index_of_restaurant(corr_mat, restaurant_input):
    restaurant_names = ratings_crosstab.columns
    restaurant_list = list(restaurant_names)

    restaurant = restaurant_list.index(restaurant_input)
    corr_rest = corr_mat[restaurant]
    return restaurant_names, corr_rest
```

### Recommending a Highly Correlated Restaurant

```python
def return_highly_correlated_list(restaurant_names, corr_rest, lr_bound=0.90):
    return list(restaurant_names[(corr_rest < 1.0) & (corr_rest > lr_bound)])
```

```python
restaurant_input = input("Enter a Restaurant: ")
ratings_crosstab = utility_matrix()
X = transpose_matrix(ratings_crosstab)
resultant_matrix = svd_fit_transform(X)
corr_mat = corr_matrix(resultant_matrix)
restaurant_names, corr_rest = get_index_of_restaurant(corr_mat, restaurant_input)
print(return_highly_correlated_list(restaurant_names, corr_rest, lr_bound=0.90))
```
